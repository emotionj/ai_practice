```python
#검거율 heatmap
#대표값을 설정한 '검거' 컬럼 기준으로 정렬

def drawGraph():
    
    #데이터 프레임 생성
    target_col=['살인검거율', '강도검거율', '강간강제추행검거율', '절도검거율', '폭력검거율','검거']

    crime_anal_norm_sort=set_crime_norm.sort_values(by='검거',ascending=False) #내림차순 정렬

    #그래프 설정
    plt.figure(figsize=(10,10))
    sns.heatmap(
        data=crime_anal_norm_sort[target_col],
        annot=True,
        fmt='.2f', #.0f: 정수, .2f: 소수점2째자리까지
        linewidths=0.5, #간격 설정
        cmap='RdPu',
        
    )
    plt.title('범죄 검거비율(정규화된 검거 합으로 정렬)')
    plt.show()
```

```python
drawGraph()
```

#### 결과해석
1. 구별 강도검거율은 주로 높다
2. 구별 절도검거율은 주로 낮다

```python
#범죄 발생 건수 heatmap

#'범죄' 컬럼 기준으로 정렬

def drawGraph():
    
    #데이터 프레임 생성
    target_col=['살인', '강도', '강간강제추행', '절도', '폭력','범죄']

    crime_anal_norm_sort=set_crime_norm.sort_values(by='범죄',ascending=False) #내림차순 정렬

    #그래프 설정
    plt.figure(figsize=(10,10))
    sns.heatmap(
        data=crime_anal_norm_sort[target_col],
        annot=True,
        fmt='.2f', #.0f: 정수, .2f: 소수점2째자리까지
        linewidths=0.5, #간격 설정
        cmap='RdPu',
        
    )
    plt.title('범죄비율(정규화된 검거 합으로 정렬)')
    plt.show()
```

```python
drawGraph()
```

#### 결과해석
1. 강남구 : 범죄 전체 1등
2. 도봉구 : 범죄비율이 대체적으로 제일 낮다. 제일 안전한 지역인걸까?
3. 범죄순위 : 강남구 > 송파구 > 관악구 

```python
#cctv 수 heatmap

#'cctv' 컬럼 기준으로 정렬

def drawGraph():
    
    #데이터 프레임 생성
    target_col=['살인', '강도', '강간강제추행', '절도', '폭력','범죄', 'CCTV']

    crime_anal_norm_sort=set_crime_norm.sort_values(by='CCTV',ascending=False) #내림차순 정렬

    #그래프 설정
    plt.figure(figsize=(10,10))
    sns.heatmap(
        data=crime_anal_norm_sort[target_col],
        annot=True,
        fmt='.2f', #.0f: 정수, .2f: 소수점2째자리까지
        linewidths=0.5, #간격 설정
        cmap='RdPu',
        
    )
    plt.title('범죄비율(정규화된 검거 합으로 정렬)')
    plt.show()
```

```python
drawGraph()
```

#### 결과해석
1. CCTV 수가 가장 높은곳은 강남구 이지만 범죄비율또한 강남구가 높다
2. 범죄3위 관악구 또한 CCTV 수가 높지만 범죄비율이 높은편이다
3. 도봉구는 CCTV 수가 적은 편이고, 범죄비율 또한 낮다

! CCTV 수가 적다고 해서 범죄가 많이 일어나는 것은 아니다. 방범용 CCTV를 더 설치해야하는걸까?
