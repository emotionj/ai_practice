```python


#절대값 알고리즘 (부호판단)
#양수 >> 양수, 음수 >> 양수로 변경(-) 곱해서 출력됨


#절대값 알고리즘 1 (부호판단)
#입력:실수 a
#출력:a의 절대값

def abs_sign(a):
  if a>=0:
    return a
  else:
    return -a


#절대값 알고리즘 2 (제곱-제곱근)
#입력:실수 a
#출력:a의 절대값

def abs_square(a):
  b=a*a
  return math.sqrt(b) #수학 모듈의 제곱근 함수

print(abs_sign(5))
print(abs_sign(-3))
print()
print(abs_square(5))
print(abs_square(-3))


```

```python
#최대값 구하는 알고리즘

#최대값 구하기
#입력 : 숫자가 N개 들어가 있는 리스트
#출력 : 숫자 N개 중 최대값

def find_max(a):
  n=len(a) #입력 크기 n
  max_v=a[0] #리스트의 첫번째 값을 최대값이라 전제
  for i in range(1,n): #1부터 n-1 까지 반복
    if a[i] > max_v: #이번 값이 현재까지 기억된 최대값 보다 크면
      max_v = a[i] #최대값 변경
  return max_v

a=[17,92,18,33,58,7,33,42]
print(len(a))
print(find_max(a))




```

```python
#최대값의 위치를 구하는 알고리즘

#최대값의 위치 구하기
#입력 : 숫자 n개 들어가 있는 리스트
#출력 : 숫자 n개 중에서 최대값이 있는 위치 (0~n-1 값)


def find_max_idx(a):
  n=len(a)
  max_idx=0 #리스트 중 0번 위치를 최대값 위치로 기억(전제)
  for i in range(1,n):
    if a[i] > a[max_idx]: #이번값이 현재까지 기억된 최대값보다 크면
      max_idx=i #최대값의 위치를 변경
  return max_idx

print(find_max_idx(a))
```

### 동명이인 찾기


```python
#두 번 이상 나온 이름 찾기
#입력: 이름이 n개 들어 있는 리스트
#출력: 이름이 n개 중 반복되는 이름의 집합

def find_same_name(a):
  n=len(a) #리스트의 자료 개수를 n에 저장
  result=set() #set():중복 제거, 결과를 저장한 빈 집합

  for i in range(0,n-1): # 마지막은 비교할 대상이 없기때문에, 0부터 n-2까지 반복
    for j in range(i+1,n): #i+1 부터 n-1까지 비교(두번째 이름부터 비교해야해서)
      if a[i] == a[j]: #이름이 같다면
        result.add(a[i])
  return result
```

```python
a=['tom','jerry','mike','tom']
print(find_same_name(a))
```

```python
name=['tom','jerry','mike','tom','jerry',',ming']
print(find_same_name(name))
```

### 팩토리얼 구하기

```python
#연속한 숫자의 곱을 구하는 알고리즘
#입력: n
#출력: 1부터 n까지 연속한 숫자를 곱한 값

def factorial(n):
  f=1 #초기값 설정, 초기값 1 설정(곱을 계산할 변수)
  for i in range(1,n+1): #1부터 n까지 반복(n+1 제외)
    f=f*i
  return f
```

```python
print(factorial(1))
print(factorial(5))
print(factorial(10))
```

```python
#연습문제 풀이
#재귀 호출 이용, 1부터 n까지 합 구하기

def sum_n(n):
  if n == 0:
    return 0
  return sum_n(n-1)+n
```

```python
#연습문제 풀이
#재귀 호출 이용, 최대값 구하기

def find_max(a,n):#리스트 a의 앞부분 n개 중 최대값을 구하는 재귀함수
  if n==1:
    return a[0]
  max_n_1=find_max(a,n-1) #n-1 개 중 최대값 구함
  if max_n_1 > a[n-1]: #n-1개 중 최대값과 n-1번 위치 값 비교
    return max_n_1
  else:
    return a[n-1]
```

```python
a=[17,92,18,33,58,7,33,42]
print(find_max(a,len(a)))
```

### 최대공약수 알고리즘

```python
#재귀 호출을 이용한 최대값 찾기
#최대 공약수 구하기

#입력 : a,b (정수)
#출력 : a와 b의 최대공약수
```

```python
from re import I
def gcd(a,b): #a,b 두개의 정수 입력값으로 받음
  i=min(a,b) #a,b 중 가장 작은 값을 i 에 저장
  #예 a=4 b=6 >> i=4
  while True:
    if a % i ==0 and b%i ==0:
      return i

    # 4/4 나머지(0) and(동시만족) 6/4 나머지(2) >> 조건만족안함
    # 4/3 나머지(0)
    # 4/2 나머지(0) and 6/2 나머지(0) >> 조건만족
    i = i-1
#4>>3>>2 : 최대공약수 2
```

```python
print(gcd(4,6))
```

```python
print(gcd(60,24))
```

```python
#최대 공약수
#입력: a,b
#출력: a와b의 최대 공약수
```

```python
def gcd(a,b):
  if b == 0: #종료조건
   return a
  return gcd(b, a%b) #좀 더 작은 값으로 자기 자신을 호출
```

```python
gcd(1,5)
```

```python
gcd(3,0)
```

```python
def gcd(a,b):
  print('gcd:',a,b) #함수 호출을 입력한 인자 값과 함께 화면에 표시하고자 함
  if b==0:
    return a
  return gcd(b, a%b)
```

```python
#n번째 피보나치 수열 찾기
#입력: n값(0부터 시작)
#출력: n번째 피보나치 수열 값

def fibo(n):
  if n <=1:
    return n #n=0>>0, n=1>>>1
  return fibo(n-2)+fibo(n-1)
```

```python
print(fibo(7))
#n번 피보나치수 = n-2번 피보나치 수 + n-1번 피보나치 수
```
