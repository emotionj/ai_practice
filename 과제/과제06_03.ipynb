```python
#쉽게 설명한 퀵 정렬

#입력: 리스트 a
#출력: 정렬된 새 리스트
```

```python
def quick_sort(a):
  n=len(a)
  if n <= 1:
    return a #종료조건
  #기준값 정하고, 기준에 맞추어 그룹을 나누기
  pivot=a[-1] #리스트의 마지막 값을 기준 값으로 설정(임의)
  g1=[] #기준 값 보다 작은 값을 담을 리스트
  g2=[] #기준 값 보다 큰 값을 담을 리스트

  for i in range(0,n-1): #마지막 값(a[-1]) 기준값으로 제외
    if a[i] < pivot: #기준값(pivot) 과 비교해서 작으면 g1에 추가
      g1.append(a[i])
    else:
      g2.append(a[i]) #크면 g2에 추가

  return quick_sort(g1)+[pivot]+quick_sort(g2)
```

```python
#일반화 된 퀵정렬

#입력: 리스트 a
#출력: 입력으로 주어진 a가 정렬
#리스트 a의 어디서부터(start) 어디까지(end)가 정렬대상인지 범위 지정
#이를 정렬하는 재귀 호출 함수
```

```python
def quick_sort_general(a,start,end):
  if end-start<=0:
    return
  pivot=a[end] #리스트의 마지막 값을 기준값으로 설정
  i=start
  for j in range(start,end):
    if a[j] <= pivot:
      a[i], a[j]=a[j], a[i]
      i += 1
  a[i], a[end] = a[end], a[i]

  #재귀 호출
  quick_sort_general(a, start, i-1) #기준 값보다 작은 그룹 >> 재귀호출로 재 정렬
  quick_sort_general(a, i+1, end) #기준 값보다 큰 그룹 >> 재귀호출 재 정렬

#리스트 전체(0~len(a)-1)를 대상으로 재귀 호출 함수 호출
def quick_sort(a):
  quick_sort_general(a, 0, len(a)-1)
```

```python
#거품 정렬(bubble sort)

#입력: 리스트 a
#출력: 입력으로 주어진 a가 정렬됨
```

```python
def bubble_sort(a):
  n=len(a)
  while True: #정렬이 완료될 때까지 계속 수행
    changed=False #자료를 앞뒤로 바뀌었는지 여부
    #자료를 훑어 보면서 뒤집힌 자료가 있는지 확인하면서 있으면 바꾸고 바꾼것을 표시해달라
    for i in range(0,n-1):
      if a[i]>a[i+1]: #앞이 뒤보다 크면
        print(a)
        a[i], a[i+1] = a[i+1], a[i] #두 자료의 위치를 맞바꿈
        changed=True #두 자료가 앞뒤로 바뀌었음을 표시

    if changed==False: #바뀐적이 없다면 정렬 완성된 것으로 종료
      return
```

```python
#리스트에서 특정 숫자 위치 찾기(이분 탐색)
#입력: 리스트 a, 찾는 값 x
#출력: 찾으면 그 값의 위치, 찾지 못하면 -1

#탐색할 범위를 저장하는 변수 start, end
#리스트 전체를 범위로 탐색 시작(0~len(a)-1)
```

```python
def binary_search(a,x):
  start=0
  end=len(a)-1

  while start<=end: #탐색할 범위가 남아 있는 동안 반복
    mid=(start+end)//2 #탐색 범위의 중간 위치
    if x == a[mid]:
      return mid
    elif x> a[mid]: #찾는 값이 더 크면 >> 오른쪽으로 범위를 좁혀 계속 탐색
      start=mid+1
    else:
      end=mid-1
  return -1 #찾지 못했을 때
```

```python
#리스트에서 특정 숫자 위치 찾기(이분 탐색과 재귀 호출)
#입력: 리스트 a, 찾는 값 x
#출력: 찾으면 그 값의 위치, 찾지 못하면 -1

#리스트 a의 어디서부터(start) 어디까지(end)가 탐색 범위인지 지정하여 그 범위 안에서 x를 찾는 재귀함수
```

```python
def binary_search_sub(a,x,start,end):
  if start > end:
    return -1

  mid=(start+end)//2  #탐색범위의 중간 위치
  
  if x==a[mid]:
    return mid
  elif x>a[mid]:
    return binary_search_sub(a,x,mid+1,end)
  else:    
    return binary_search_sub(a,x,start,mid-1)
  
  return -1
```

```python
#리스트 전체(0~len(a)-1)를 대상으로 재귀 호출 함수 호출

def binary_search(a,x):
  return binary_search_sub(a,x,0,len(a)-1)
```

```python
#회문찾기 알고리즘
#주어진 문장(s)이 회문인지 아닌지 판별(queue, stack 특징 활용)

#입력: s(문자열)
#출력: 회문 >> True, False
```

```python
def palindrome(s):
  qu=[] #queue 초기화
  st=[] #stack 초기화

  #1.문자열의 알파벳 문자를 각각 큐와 스택에 담는다
  for x in s:
    if x.isalpha():
      qu.append(x.lower())
      st.append(x.lower())
  #2. 큐와 스택에 들어있는 문자를 꺼내면서 비교
  while qu: #queue 에 문자가 남아 있는 동안은 계속 반복
    if qu.pop(0) != st.pop(): #큐와 스택에서 꺼낸 문자가 다르다면 >> 회문 아님
      return False
  return True
```

```python
#문자열 앞뒤를 서로 비교, 회문인지 확인

#입력: 문자열 s
#출력: 회문>True,False
```

```python
def palindrome(s):
  i=0 #문자열 s의 앞에서 비교할 위치
  j=len(s)-1

  while i<j: #중간까지 검사하기
    #i위치에 있는 문자가 알파벳 문자가 아니면 뒤로 이동
    if s[i].isalpha()==False:
      i+=1 #i=i+1
    #j위치에 있는 문자가 알파벳 문자가 아니면 앞으로 이동
    elif s[j].isalpha()==False:
      j-=1 #j=j-1
      #i,j 위치 둘 다 알파벳 문자가 있으면 두 문자 비교, 다르면 회문이 아님
    elif s[i].lower() != s[j].lower():
      return False
    
    else:
      i+=1
      j-=1
  return True

```

```python
#딕셔너리를 이용한 동명이인 찾는 알고리즘

#두 번 이상 나온 이름 찾기

#입력: 이름이 n개 들어 있는 리스트 >> a
#출력: n개 이름 중에서 반복되는 이름의 집합
```

```python
def find_samename(a):
  #1단계: 각 이름의 등장 횟수를 dict() 만들어줘
  name_dict={}
  for name in a: #리스트 a에 있는 data를 차례로 반복
    if name in name_dict: #이름이 name_dict에 있으면
      name_dict[name] += 1 #등장 횟수를 1 증가
    else:
      name_dict[name]=1 #등장횟수에 1로 저장
    
  #2단계: 만들어진 dict()에서 등장 횟수가 2이상인 것을 결과에 추가
  result=set() #결과값을 저장한 빈 집합
  for name in name_dict:
    if name_dict[name] >= 2:
      result.add(name)
    return result
```

```python
name=['Tom','Jerry','Drwill','Zzuyu', 'Kazuha']
print(find_samename(name))
```

```python
name=['Drwill','Drwill','Zzuyu', 'Kazuha']
print(find_samename(name))
```

```python
def get_name(s_info,find_no):
  if find_no in s_info:
    return s_info[find_no]
  else:
    return "?" #해당하는 find_no (번호)가 없으면 물음표
```

```python
#친구 리스트에서 자신의 모든 친구를 찾는 알고리즘

#입력: 친구관계 그래프 g, 모든 친구를 찾을 자신 start
#출력: 모든 친구의 이름

def print_all_friends(g,start):
  qu=[] #기억 장소1: 앞으로 처리해야 할 사람들을 큐에 저장
  done=set() #기억 장소2: 이미 큐에 추가한 사람들을 집합에 기록(중복 방지)

  qu.append(start) #자신을 큐에 넣고 시작
  done.add(start) #집합에도 추가

  while qu: #큐에 처리할 사람이 남아 있는 동안
    p=qu.pop(0) #큐에서 처리 대상을 한 명 꺼내
    print(p) #그 이름을 출력하고
    for x in g[p]: #그의 친구들 중에
      if x not in done: #아직 큐에 추가된 적이 없는 사람을
        qu.append(x) #큐에 추가하고
        done.add(x) #집합에도 추가

#친구 관계 리스트
#A와 B가 친구이면
#A의 친구 리스트에도 B가 나오고, B의 친구 리스트에도 A가 나옴

fr_info={
    'Summer':['John','Justin','Mike'],
    'John':['Summer','Justin'],
    'Justin':['John','Summer','Mike','May'],
    'Mike':['Summer','Justin'],
    'May':['Justin','Kim'],
    'Kim':['May'],
    'Tom':['Jerry'],
    'Jerry':['Tom']
}

print_all_friends(fr_info,'Summer')
print()
print_all_friends(fr_info,'Jerry')
```

```python
#친구 리스트에서 자신의 모든 친구를 찾고 친구들의 친밀도를 계산하는 알고리즘

#입력: 친구 관계 그래프 g, 모든 친구를 찾을 자신 start
#출력: 모든 친구의 이름과 자신과의 친밀도
```

```python
def all_friends(g, start):
  qu=[] #기억장소 1: 앞으로 처리해야 할 (사람이름, 친밀도) 튜플을 큐에 저장
  done=set() #기억장소 2: 이미 큐에 추가한 사람을 집합에 기록(중복 방지)
  qu.append((start,0)) #((사람이름,친밀도)) 하나의 튜플로 묶음
  #start:자기자신, 친밀도:0
  done.add(start)

  while qu: #큐에 처리할 사람이 남아 있는 동안
    (p,d)=qu.pop(0) #큐에서 (사람이름, 친밀도) 정보를 p와 d로 각각 꺼냄
    print(p,d)
    for x in g[p]: #친구들 중에서
      if x not in done: #아직 큐에 추가된 적이 없는 사람을
        qu.append((x,d+1)) #친밀도(d)를 1증가시켜 큐에 추가
        done.add(x)

fr_info={
    'Summer':['John','Justin','Mike'],
    'John':['Summer','Justin'],
    'Justin':['John','Summer','Mike','May'],
    'Mike':['Summer','Justin'],
    'May':['Justin','Kim'],
    'Kim':['May'],
    'Tom':['Jerry'],
    'Jerry':['Tom']
}
```

```python
all_friends(fr_info,'Summer')
```

```python
all_friends(fr_info,'Jerry')
```

```python
#그래프 탐색: 너비 우선 탐색

#입력: 그래프 g, 탐색 시작점 start
#출력: start에서 출발해 연결된 꼭짓점들을 출력

def bfs(g, start):
  qu=[] #기억장소1: 앞으로 처리해야 할 꼭짓점을 큐에 저장
  done=set() #기억장소2: 이미 큐에 추가한 꼭짓점들을 집합에 기록(중복방지)

  qu.append(start) #시작점을 큐에 넣고 시작
  done.add(start) #집합에도 추가

  while qu: #큐에 처리할 꼭짓점이 남아있으면
    p=qu.pop(0) #큐에서 처리 대상을 꺼내어
    print(p) #꼭짓점 이름을 출력하고
    for x in g[p]: #대상 꼭짓점에 연결된 꼭짓점들 중에
      if x not in done: #아직 큐에 추가된 적이 없는 꼭짓점들을
        qu.append(x) #큐에 추가하고
        done.add(x) #집합에도 추가

#그래프 정보
g={
  1:[2,3],
  2:[1,4,5],
  3:[1],
  4:[2],
  5:[2]
}

bfs(g,1)
```
