```python
# 하노이의 탑

# 입력: 옮기려는 원반의 개수 n
# from_pos: 옮길 원반이 현재 있는 출발점 기둥
# to_pos: 원반을 옮길 도착점 기둥
# aux_pos: 옮기는 과정에서 사용할 보조 기둥

# 출력: 원반을 옮기는 순서
```

```python
def hanoi(n, from_pos, to_pos, aux_pos):
  if n==1: #원반 1개 >> 그냥 옮기면 된다
    print(from_pos, '->', to_pos)
    return

  #원반 n-1개 >> aux_pos로 이동(to_pos 보조기둥으로 이용)
  hanoi(n-1, from_pos, aux_pos, to_pos)
  #가장 큰 원반을 목적지로 이동
  print(from_pos,'->',to_pos)
  #aux_pos에 있는 n-1개를 목적지로 이동(from_pos 보조기둥으로 사용)
  hanoi(n-1,aux_pos, to_pos, from_pos)
```

```python
print('n=1')
hanoi(1,1,3,2)
#원반 1개, 1번 기둥에서 3번 기둥으로 이동 (2번 기둥: 보조기둥)
```

### 쉽게 설명한 선택 정렬 알고리즘

```python
# 입력: 리스트 a
# 출력: 정렬된 새 리스트
```

```python
def find_min_idx(a):
  n=len(a)
  min_idx=0
  for i in range(1,n):
    if a[i]<a[min_idx]:
      min_idx=i
  return min_idx
```

```python
def sel_sort(a):
  result = [] #새 리스트 만들어 정렬된 값을 저장
  while a:#주어진 리스트에 값이 남아 있는 동안 계속 반복
    min_idx=find_min_idx(a) #리스트에 남아 있는 값 중에서 최소값 위치
    value=a.pop(min_idx) #찾은 최소값 빼내어 value 저장
    result.append(value) #value를 결과(result) 리스트 끝에 추가
  return result
```

### 쉽게 설명한 삽입 알고리즘

```python
# 입력: 리스트 a
# 출력: 정렬된 새 리스트
```

```python
# 함수 정의
# 리스트 r(result)에서 v(value)가 들어가야 할 위치를 돌려주는 함수

def find_ins_idx(r,v):
  for i in range(0,len(r)):
    if v < r[i]: #v값 보다 i번 위치에 있는 자료 값이 크면
      return i #v가 그 값 바로 앞에 놓이도록 함 (정렬 순서가 유지돔)
  return len(r) #적절한 위치를 못 찾았을 때는 v가 r의 모든 자료보다 크다는 의미
```

```python
def ins_sort(a):
  result=[] #새 리스트를 생성, 정렬된 값을 저장
  while a: #기존 리스트에 값이 남아 있는 동안 계속 반복
    value=a.pop(0) #기존 리스트에서 1개를 꺼냄
    ins_idx = find_ins_idx(result, value)
    result.insert(ins_idx,value)
  return result
```

```python
#입력: 리스트 a
#출력: 입력으로 주어진 a가 정렬됨
def ins_sort(a):
  n=len(a)
  for i in range(1,n): #1부터 n-1까지
    #i번 위치의 값을 key로 저장
    key=a[i]
    #j를 i 바로 왼쪽 위치로 저장
    j=i-1
    while j >= 0 and a[j] > key: #key=a[i]
    #리스트의 j번 위치 있는 값과 key(=a[i]) 비교
    #key가 삽일 될 적절한 위치를 찾음
      a[j+1]=a[j] #공간 만들어 주기(삽입할 공간이 생기도록 값을 오른쪽으로 한 칸 이동)
      j -= 1 #j=j-1
    a[j+1]=key #찾은 삽입 위치에 key를 저장
```

### 쉽게 설명한 병합정렬

```python
#입력: 리스트 a
#출력: 정렬된 새 리스트

def merge_sort(a):
  n=len(a)
  if n<=1:
    return a
  #정지조건 : 정렬할 리스트의 자료 개수가 1개 이하이면 정렬 필요 없음

  #그룹을 나누어 각각 병합 정렬 호출
  mid=n//2 #중간기준, 2개의 그룹으로 나눔
  g1=merge_sort(a[:mid]) #재귀호출로 1번째 그룹 정렬
  g2=merge_sort(a[mid:]) #재귀호출로 2번째 그룹 정렬

  #2개의 그룹 하나로 병합
  result=[] #최종결과물 담을 빈 리스트 생성
  while g1 and g2:
    if g1[0] < g2[0]: #2개의 그룹 맨 앞 자료값 비교
    #g1의 값이 더 작으면 >> 그 값을 빼내어서 결과로 추가
      result.append(g1.pop(0))
    else:
      result.append(g2.pop(0))

  #아직 남아있는 자료들을 결과에 추가
  #g1과 g2 중 이미 빈것은 while을 바로 지나감

  while g1:
    result.append(g1.pop(0))
  while g2:
    result.append(g2.pop(0))
  return result
```

```python
a=[6,8,3,9,10,1,2,4,7,5]
len(a)
```

```python
merge_sort(a)
```

```python
#입력: 리스트 a
#출력: 입력으로 주어진 a가 정렬됨

def merge_sort(a):
  n=len(a)
  if n<=1:
    return 
  
  mid=n//2 #중간을 기준
  g1=a[:mid]
  g2=a[mid:]
  merge_sort(g1)
  merge_sort(g2)

  #2그룹 >> 1그룹 병합
  i1=0
  i2=0
  ia=0
  while i1<len(g1) and i2<len(g2):
    if g1[i1]<g2[i2]:
      a[ia]=g1[i1]
      i1+=1 #i1=i1+1
      ia+=1
    else:
      a[ia]=g2[i2]
      i2+=1
      ia+=1
  #아직 남아 있는 자료들을 결과에 추가
  while i1<len(g1):
      a[ia]=g1[i1]
      i1+=1
      ia+=1
  while i2<len(g2):
      a[ia]=g2[i2]
      i2+=1
      ia+=1

```
