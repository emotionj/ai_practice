matplotlib의 플롯은 `Figure`라는 객체 안에 위치한다. 새로운 Figure는 다음과 같이 만들 수 있다.

```python
fig=plt.figure()
```

Figure에 실제 그래프를 넣기 위해서는 subplot이 필요하다. 

    ax1 = fig.add_subplot(2, 2, 1)

위 코드는 figure가 2 × 2 모양이며 우리는 그 중에 1번째 subplot을 선택한다는 뜻이다.

(matplotlib의 인덱스는 1부터 시작한다. 파이썬 표준을 따르지 않는 이유는 matplotlib이 만들어질 때 matlab을 본따 만들어졌기 때문이다.)

```python
ax1=fig.add_subplot(2, 2, 1)
```

명령어들을 한 셀에서 실행해 보자.

```python
fig=plt.figure()
ax1=fig.add_subplot(2, 2, 1)
ax2=fig.add_subplot(2, 2, 2)
ax3=fig.add_subplot(2, 2, 3)
```

#### 판다스와 matplotlib

판다스에서는 `.plot()`이라는 간단한 메서드로 데이터를 matplotlib으로 시각화하는 방법을 제공한다.

    s.plot()

```python
s=pd.Series(np.random.randn(10).cumsum(),index=np.arange(0,100,10))
```

```python
s
```

```python
s.plot()
```

데이터프레임에서도 `.plot()` 메서드가 존재한다. 데이터프레임의 plot은 여러 컬럼의 값을 동시에 보여줄 수 있다.

    df.plot()

```python
df=pd.DataFrame(np.random.randn(10,4).cumsum(0),columns=['a','b','c','d'],index=np.arange(0,100,10))
```

```python
df.plot()
```

바 차트는 다음과 같이 만들 수 있다.

```python
fig,axes = plt.subplots(2, 1)
data=pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))

data.plot.bar(ax=axes[0], color='k', alpha=0.7)
data.plot.barh(ax=axes[1], color='k', alpha=0.7)
```

데이터프레임에서 바 차트는 자동으로 그룹화되어 나타난다.

# 인터렉티브 시각화

bokeh와 plotly의 가장 큰 특징은 인터렉티브 시각화를 생성한다는 것이다. matplotlib이 인쇄물에 최적화되어 있는 반면에 bokeh와 plotly는 브라우저 환경에 특화되어 있다.

#### bokeh


```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

#노트북에서 bokeh 플롯을 출력하기 위해 실행
output_notebook()
```

```python
N=4000
x=np.random.random(size=N)*100
y=np.random.random(size=N)*100
radii=np.random.random(size=N)*1.5
colors=["#%02x%02x%02x" % (r, g, 150) for r, g in zip(np.floor(50+2*x).astype(int), np.floor(30+2*y).astype(int))]

p=figure()
p.circle(x, y, radius=radii, fill_color=colors, fill_alpha=0.6, line_color=None)
show(p)
```

import plotly.express as px
df = px.data.iris()

```python
import plotly.express as px
df=px.data.iris()
fig=px.scatter(df, x="sepal_width", y="sepal_length", color="species")
fig.show()
```

다음과 같이 복잡한 그래프도 그릴 수 있다.

```python
import plotly.express as px
df=px.data.iris()
fig=px.scatter(df, x="sepal_width", y="sepal_length", color="species", marginal_y="violin",
              marginal_x="box", trendline="ols", template="simple_white")
# x축 가장자리: 바이올린 그래프 
# y축 가장자리: box plot 
# ols : 회귀분석 >> trendline (회귀선) 그려줌
fig.show()
```
