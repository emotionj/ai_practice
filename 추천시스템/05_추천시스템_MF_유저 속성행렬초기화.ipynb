```python
np.random.seed(32)

#임의로 유저 취향과 상품 속성 행렬 생성하는 함수
def initialize(R,num_features):
  #유저 취향: 유저 데이터 개수 * 속성개수차원(k) 임의 행렬(random) >> 변수 theta 저장
  #상품속성 취향 : 속성개수차원(k) 임의행렬(random) * item(영화) 데이터 개수 차원(random) >> 변수 X 저장
  num_users, num_items=R.shape

  Theta=np.random.rand(num_users,num_features)
  X=np.random.rand(num_features,num_items)

  return Theta,X
```

```python
#실제 값 행렬
R=np.array([
    [3,4,1,np.nan,1,2],
    [4,4,3,np.nan,5,3],
    [2,3,np.nan,1,3,4],
    [1,3,2,4,2,2],
    [1,2,np.nan,2,2,4]
])
```

```python
Theta,X=initialize(R,2)
#초기화한 값
```
