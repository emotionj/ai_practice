```python
#CBF 내용기반 추천시스템
from sklearn.linear_model import LinearRegression #회귀분석
from sklearn.model_selection import train_test_split

import numpy as np
import pandas as pd
```

```python
#유저 평점 + 영화 속성 데이터 경로 정의
#/content/drive/MyDrive/Classroom/인공지능 서비스 개발자 _ 강동1기/data/movie_rating.csv

movie_data_path='/content/drive/MyDrive/Classroom/인공지능 서비스 개발자 _ 강동1기/data/movie_rating.csv'
```

```python
#판다스로 데이터 불러오기

pd.read_csv(movie_data_path)
```

```python
df=pd.read_csv(movie_data_path)
df[:5]
```

```python
df.columns
```

```python
features=['romance', 'action', 'comedy', 'heart-warming']
```

```python
#입력변수와 목표변수(결과변수 - rating) 나누기

x=df[features]
y=df[['rating']] #2차원 동일한 형태로 만들어줌
```

```python
x
```

```python
y
```

```python
#train, test data set 분리

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3, random_state=32)
#train 70%, test 30% ~시험 문제~
```

```python
#모델 훈련 (회귀분석 모델 가져옴)

model=LinearRegression()
model.fit(x_train,y_train)
#fit할 때는 반드시 훈련용 데이터로만 사용!
# -->> 훈련용 데이터로 모델 구축! (테스트 데이터는 건들면 안된다)
#fit 정규화! 정확한 측정을 위한 정규화! 사용할 모델 적합하게 데이터를 정규화 해줌
```

```python
#회귀분석:과거의 데이터 분포(산점도)를 보고 미래를 추정하는 것

model.predict(x_test)
#예측할 때는 테스트 데이터로 예측
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred=pd.DataFrame(y_pred)
y_pred
```
