```python
#코사인 유사도 이용, 영화별 유사도 계산
from sklearn.metrics.pairwise import cosine_similarity

genre_sim=cosine_similarity(genre_mat,genre_mat)
print(genre_sim.shape)
print(genre_sim[:5])
```

```python
#argsort:자료를 정렬하는 것이 아님. 순서만 알고 싶을 때 사용
#유사도가 높은 영화를 앞에서부터 순서대로 보여줌
#0번째 영화의 경우 유사도 순서 : 0번, 3494번, 813번 . . . . 순서

genre_sim.argsort()[:,::-1] #내림차순 정리, 유사도가 높은것 >> 낮은것
# -->> 개별 영화 끼리의 유사도

genre_sim_sorted_ind=genre_sim.argsort()[:,::-1]
```

```python
#실제 함수
#sorted_ind,title_name 은 변수 설정
def find_sim_movie_ver1(df,sorted_ind,title_name,top_n=10):
  #인자로 입력된 df 데이터프레임에서 'title' 컬럼이 입력된 title_name 값인 데이터프레임 추출
  title_movie=df[df['title']==title_name]
  #title_name 을 가진 데이터프레임의 index 객체를 ndarray 로 반환하고
  #sorted_ind 인자로 입력된 genre_sim_sorted_ind 객체에서 유사도 순으로 top_n개의 index 추출

  title_index=title_movie.index.values
  similar_indexes=sorted_ind[title_index,:(top_n)]

  #추출된 top_n index를 출력(2차원)
  #데이터프레임에서 index로 사용하기 위해서는 1차원 array로 변경

  print(similar_indexes)

  #2차원 >> 1차원
  similar_index=similar_indexes.reshape(-1)

  return df.iloc[similar_index]
```

```python
##(매우 중요) CBF 활용시 적용해야 할 코드

#가중평점(평점,평가횟수) 반영한 영화 추천
```

```python
# 상위 60% 에 해당하는 vote_count 최소 투표 횟수인 m 으로 지정
df['vote_average']
df['vote_average'].mean()

C=df['vote_average'].mean()
print(C)
#전체 영화에 대한 평점 6.092 

df['vote_count'].quantile(0.6)
m=df['vote_count'].quantile(0.6)
print(m)
#370.1999999999998 -->> 적어도 370번 이상 평점을 준 영화의 평점을 사용(투표수 370번 이상)
```

```python
#C:전체 영화에 대한 평균 평점 : 약 6점
#m:평점을 부여하기 위한 최소 투표 회수 = 370회(상위 60% 수준)

print('C: ',round(C,2),'점', 'm: ',round(m,2),'회')
```

```python
#가중 평점을 계산하는 함수
def weighted_vote_average(record):
  v=record['vote_count']
  R=record['vote_average']

  return (((v/(v+m))*R) + ((m/(m+v))*C))
```

```python
#기존 데이터에 가중평점 컬럼 추가

df.apply(weighted_vote_average, axis=1)
df['weighted_vote']=df.apply(weighted_vote_average, axis=1)
```

```python
#가중평균 기준, 영화 순위 추출(내림차순)>>평점이 높은 영화 먼저 출력

df.columns
```

```python
df[['weighted_vote','title','vote_average','vote_count','genres']]
df[['weighted_vote','title','vote_average','vote_count','genres']].sort_values('weighted_vote', ascending=False)[:10]
```

```python
##추천 ver2 먼저 장르 유사성 높은 영화 20개 선정 후, 가중평점 순으로 10개 선정

def find_sim_movie_ver2(df, sorted_ind, title_name, top_n=10):
  title_movie= df[df['title']==title_name]
  title_index=title_movie.index.values

  #top_n의 2배 (유사성 높은 영화 20개 선정 위함)에 해당하는 장르 유사성이 높은 index 추출
  similar_indexes=sorted_ind[title_index, :(top_n*2)]
  similar_index=similar_indexes.reshape(-1)

  #기준 영화 index는 제외(기존에 본 영화 제외하고 추천하기 위함. title_index 겹치지않는영화 추천)
  similar_index=similar_index[similar_index != title_index]

  #top_n의 2배(후보군)에서 weighted_vote 높은 순으로 top_n 만큼 추출
  return df.iloc[similar_index].sort_values('weighted_vote', ascending=False)[:top_n]



```

```python
#영화 The Godfather에 대한 장르 유사성, 가중평점 반영한 영화 10편을 추천
similar_movies=find_sim_movie_ver2(df, genre_sim_sorted_ind, 'The Godfather',10)
similar_movies
```
