```python
def distance(user_1,user_2):
  return np.sqrt(np.sum((user_1-user_2)**2))
```

```python
#movie_id 번째 영화를 평가하지 않은 사용자 미리 제외하는 함수 생성
def filter_users_without_movie(rating_data,movie_id):
  return rating_data[~np.isnan(rating_data[:,movie_id])]
```

```python
#평점 데이터의 빈 값을 각 사용자(유저) 평균값으로 채우는 함수 생성

def fill_nan_with_user_mean(rating_data):
  #평점 데이터 원본 훼손하지 않기 위해 복사를함
  filled_data=np.copy(rating_data)
  
  #사용자(유저)의 평균 평점 계산
  row_mean=np.nanmean(filled_data,axis=1)

  #비어있는 인덱스 구하기
  inds=np.where(np.isnan(filled_data))

  #빈 인덱스를 유저 평균 평점으로 채우기
  filled_data[inds]=np.take(row_mean, inds[0])

  return filled_data
```

```python
#user_id에 해당하는 사용자(유저) 이웃을 찾아주는 함수
def get_k_neighbors(user_id,rating_data,k):
  #평점 데이터 원본 훼손하지 않기 위해 복사를함
  distance_data=np.copy(rating_data)

  #마지막에 거리 데이터 담을 열 한 개 추가
  distance_data=np.append(distance_data,np.zeros((distance_data.shape[0],1)), axis=1)

  for i in range(len(distance_data)):
    row=distance_data[i]
    
    #반복문을 돌면서 user_id번째 유저가 나올 때는 
    #해당 데이터의 거리 정보는 무한대인 np.inf로 저장함

    if i == user_id:
      #같은 유저면 거리를 무한대로 설정
      #같은 유저라도 고유위치를 갖고있을 수 있기때문에 오류가 생기지 않기위해서 inf 로 설정
      row[-1]=np.inf      
    else:
      #같은 유저가 아니라면 끝자리에 넣어줘
      #distance 는 거리구하는 함수(위에서 정의)
      row[-1]=distance(distance_data[user_id][:-1],row[:-1])

  #데이터를 거리 열 기준 정렬, argsort 하는 순간 내림차순으로 정렬
  distance_data[np.argsort(distance_data[:,-1])]

  #가장 가까운 k개의 행 리턴
  #단, 마지막(거리) 열은 제외한다
  return distance_data[:k,:-1]
```

```python
#실행 코드
#영화 3을 본 유저들 중, 유저 0과 비슷한 유저를 5명 찾으세요.

#1.평점 데이터를 불러옴
#index_col 인덱스를 user_id으로 지정
pd.read_csv(path,index_col='user_id')
rating_data=pd.read_csv(path,index_col='user_id').values
print(rating_data)
print()

#2. 3번째 영화를 보지 않은 유저를 데이터에서 미리 제외시킨다
filltered_data=filter_users_without_movie(rating_data,3)

#3. 빈 값들이 채워진 새로운 행렬 생성
filled_data=fill_nan_with_user_mean(filltered_data)
print(filled_data)


```

```python
#영화 3을 본 유저들 중 유저 0과 비슷한 유저를 5명 찾으세요.
get_k_neighbors(0,filled_data,5)
```
